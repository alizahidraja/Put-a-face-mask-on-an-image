```python
from imutils import face_utils
import numpy as np
import argparse
import os
import imutils
import dlib
import cv2
import matplotlib.pyplot as plt

def putonmask(image):
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
    image = cv2.imread(image)
    orig = image
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # detect faces in the grayscale image
    rects = detector(gray, 1)

    # loop over the face detections
    for (i, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)


        x,y=shape[1]
        x1,y1=shape[30]
        x2,y2=shape[16]
        x3,y3=shape[14]
        x4,y4=shape[12]
        x5,y5=shape[10]
        x6,y6=shape[9]
        x7,y7=shape[8]
        x8,y8=shape[6]
        x9,y9=shape[4]
        x10,y10=shape[2]


        #cv2.line(image,(x,y),(x2,y2), color=(10,10,10),  thickness=2)
        #cv2.line(image,(x2,y2),(x1,y1), color=(10,10,10),  thickness=2)

        pts=np.array([[x,y],[x1,y1],[x2,y2],[x3,y3],
                     [x4,y4],[x5,y5],[x6,y6],[x7,y7],
                     [x8,y8],[x9,y9],[x10,y10]])
        cv2.fillPoly(image,[pts],(10,10,10))
    return image


new=putonmask("image.jpeg")
cv2.imwrite('new.jpg', new)
Image(filename='new.jpg') 
```
